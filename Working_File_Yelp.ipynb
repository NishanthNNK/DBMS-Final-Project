```python
def load_json_to_df(filepath):
    chunks = pd.read_json(filepath, lines=True, chunksize=10000)
    df_list = []
    for chunk in chunks:
        df_list.append(chunk)
    return pd.concat(df_list, ignore_index=True)
```

```python
filepath_business = "/content/drive/My Drive/DBMS Final Project/YelpData/yelp_academic_dataset_business.json"
```

```python
df_business = load_json_to_df(filepath_business)
```

```python
df_user = load_json_to_df("/content/drive/My Drive/DBMS Final Project/YelpData/yelp_academic_dataset_user.json")
df_tip = load_json_to_df("/content/drive/My Drive/DBMS Final Project/YelpData/yelp_academic_dataset_tip.json")
df_review = load_json_to_df("/content/drive/My Drive/DBMS Final Project/YelpData/yelp_academic_dataset_review.json")
```

```python
def export_dataframe_to_csv(df, file_path):
    df.to_csv(file_path, index=False)

export_path = '/content/drive/My Drive/DBMS Final Project/YelpData/'

df_list = [
    (df_business, 'business.csv'),
    (df_checkin, 'checkin.csv'),
    (df_tip, 'tip.csv'),
    (df_review, 'review.csv'),
    (df_user, 'user.csv')
]

for dataframe, filename in df_list:
    full_path = export_path + filename
    export_dataframe_to_csv(dataframe, full_path)

print("Export completed")
```
